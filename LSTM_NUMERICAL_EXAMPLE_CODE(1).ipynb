```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def tanh(x):
    return math.tanh(x)

def lstm_predict():
  # Step 1: Initialize Parameters
    # Forget Gate
    Wf, Whf, bf = 0.5, 0.1, 0
    # Input Gate
    Wi, Whi, bi = 0.6, 0.2, 0
    # Candidate
    Wc, Whc, bc = 0.7, 0.3, 0
    # Output Gate
    Wo, Who, bo = 0.8, 0.4, 0
    # Linear Output
    Wy, by = 4.0, 0

    # Initial States
    h_prev = 0.0
    C_prev = 0.0

    # Input Sequence
    X = [1, 2, 3, 4]

    print(f"{'Step':<5} | {'Input':<5} | {'f_t':<8} | {'i_t':<8} | {'C_t':<8} | {'h_t':<8}")
    print("-" * 60)

    # Step 2: Compute LSTM Values for Each Time Step
    for t, x in enumerate(X):
        # Forget gate
        f_t = sigmoid(Wf * x + Whf * h_prev + bf)

        # Input gate
        i_t = sigmoid(Wi * x + Whi * h_prev + bi)

        # Candidate cell state
        C_tilde = tanh(Wc * x + Whc * h_prev + bc)

        # Cell state update
        C_t = (f_t * C_prev) + (i_t * C_tilde)

        # Output gate
        o_t = sigmoid(Wo * x + Who * h_prev + bo)

        # Hidden state update
        h_t = o_t * tanh(C_t)

        print(f"t={t+1:<2} | {x:<5} | {f_t:.3f} | {i_t:.3f} | {C_t:.3f} | {h_t:.3f}")

        # Update states for next iteration
        h_prev = h_t
        C_prev = C_t

    # Step 3: Final Prediction
    # Linear transformation from h_t to predicted value
    y_hat = Wy * h_prev + by
    print("-" * 60)
    print(f"Final Prediction (y_hat): {y_hat:.3f}")

if __name__ == "__main__":
    lstm_predict()
```
